```python
num_fea = [fea for fea in df.columns if df[fea].dtypes != 'O']
cat_fea = [fea for fea in df.columns if df[fea].dtypes == 'O']
print(num_fea)
print(cat_fea)
```

```python
# KDE for num var

# for items in num_df:
#    sns.kdeplot(num_df[items])
#    plt.title()
#    plt.show()

plt.figure(figsize=(15, 15))
plt.suptitle('Univariate Analysis of Numerical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(0, len(num_fea)):
    plt.subplot(5, 3, i+1)
    sns.kdeplot(x=df[num_fea[i]],shade=True, color='b')
    plt.xlabel(num_fea[i])
    plt.tight_layout()
```

```python
plt.figure(figsize=(15, 20))
plt.suptitle('scatter plot for numerical feature', fontsize=20, fontweight='bold', alpha=1, y=1)

for i in range(0, len(num_fea)):
    plt.subplot(5, 3, i+1)
    sns.scatterplot(y=num_fea[i], x=df.index, data=df , color = 'b')
    plt.tight_layout()
```

```python
# countplot for cat var excluding **app column** and some other

plt.figure(figsize=(20, 20))
plt.suptitle('Univariate Analysis of Categorical Features', fontsize=20, fontweight='bold', alpha=0.8, y=1.)

for i in range(1,4):
    
    sns.countplot(x=df[cat_fea[i]])
    plt.tight_layout()
    plt.show()
```
